```python
# Install necessary libraries

# Import libraries
from transformers import AutoTokenizer, AutoModelForCausalLM

# Function to load a model and tokenizer
def load_model(model_name):
    try:
        print(f"Loading model: {model_name}")
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForCausalLM.from_pretrained(model_name)
        return tokenizer, model
    except Exception as e:
        print(f"Failed to load model {model_name}: {e}")
        return None, None

# Dictionary to store models and tokenizers
models = {
    "stablelm-2-12b": load_model("stabilityai/stablelm-2-12b"),
    "stablelm-zephyr-3b": load_model("stabilityai/stablelm-zephyr-3b"),
    "StableBeluga2": load_model("stabilityai/StableBeluga2"),
    "stable-code-instruct-3b": load_model("stabilityai/stable-code-instruct-3b")
}

# Function to generate a response using the selected model
def chat_with_ai(messages, selected_model):
    tokenizer, model = models[selected_model]
    if tokenizer is None or model is None:
        return "The selected model could not be loaded. Please choose another model."

    # Create the input prompt
    prompt = " ".join([message["content"] for message in messages])
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate response
    outputs = model.generate(**inputs, max_length=150, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract the generated text (excluding the prompt)
    return response[len(prompt):].strip()

# Chat loop
def chat():
    print("Welcome to the AI Chatbot! Type 'exit' to end the conversation.")
    print("Available models: stablelm-2-12b, stablelm-zephyr-3b, StableBeluga2, stable-code-instruct-3b")

    messages = [{"role": "user", "content": "Who are you?"}]
    selected_model = "StableBeluga2"  # Default model

    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        # Switch models
        if user_input.lower().startswith("switch model"):
            model_name = user_input.split(" ", 2)[-1]
            if model_name in models and models[model_name][0] is not None:
                selected_model = model_name
                print(f"Switched to model: {selected_model}")
            else:
                print("Invalid model name or model not loaded. Please try again.")
            continue

        # Append user input to the messages
        messages.append({"role": "user", "content": user_input})

        # Generate and display the AI response
        ai_response = chat_with_ai(messages, selected_model)
        print(f"AI: {ai_response}")

        # Append the AI response to maintain conversation context
        messages.append({"role": "assistant", "content": ai_response})

# Run the chat
chat()
```
